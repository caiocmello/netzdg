```python
import pandas as pd
```

```python
df = pd.read_csv('tweets_NetzDG.csv')
```

```python
df = df[['tweet_id', 'user_username', 'text']]
```

```python
# prompt: if there is a url in column text, write this url in a new colum called url

import re
def extract_url(text):
  url = re.findall('http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', text)
  if url:
    return url[0]
  else:
    return None
df['url'] = df['text'].apply(extract_url)
df.head()
```

```python
# prompt: if text in column text starts with RT, create a new dataframe with this lines

df_rt = df[df['text'].str.startswith('RT')]
```

```python
# prompt: remove duplicates in column text

df.drop_duplicates(subset='text', inplace=True)
```

```python
# prompt: how to get the full original url for this shortned version 'https://t.co/UEqp15q2RE'
from tqdm import tqdm
import requests
tqdm.pandas()

def get_original_url(short_url):
  try:
    response = requests.head(short_url, allow_redirects=True)
    return response.url
  except:
    return None

df1['original_url'] = df1['url'].progress_apply(get_original_url)
df1.head()
```
