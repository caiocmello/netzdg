```python
df_short = df[['user_username', 'text', 'created_at', 'retweet_count', 'like_count']]
df_short.head()
```

```python
df_filtered.to_csv('/Users/caio.mello/Documents/NetzDG_project/NetzDG_data/cluster_linking_dataset.csv', index=False)
```

```python
import re
def extract_url(text):
  url = re.findall('http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', text)
  if url:
    return url[0]
  else:
    return None
df_filtered['url'] = df_filtered['text'].apply(extract_url)
df_filtered.head()
```

```python
df_filtered.info()
```

```python
df_filtered['modularity_class'] = df_filtered['modularity_class'].replace({
    11: 'blue',
    7: 'orange',
    3: 'purple',
    4: 'red',
    6: 'turquoise'
})
```
